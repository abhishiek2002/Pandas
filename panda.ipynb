Missing Value Handling:-  Because machine learning does not accept missing values .That's why we need to handle missing values 

1. Remove the records-> Remove row in which a none value present(use when only 1%-3% data contains missing record )
2. Fill the records

```python
import pandas as pd
import numpy as np
```

```python
# np.nan are none/missing values with help of numpy
# missing value not accept in machine learning
data = {'A':[2,5,np.nan,8,np.nan,9],
       'B':[np.nan,45,np.nan,89,63,np.nan], 
       'C':[np.nan,74,np.nan,np.nan,85,4], 
       'D':[10,20,30,40,50,60]}
data
```

```python
df2 = pd.DataFrame(data)
df2
```

```python
df2.dropna()  # by default remove all those rows which contains missing values but changes are not allow without inplace function
```

```python
df2.dropna(axis='columns')   #drop column having missing value
```

```python
# to check where missing value exist
df2.isnull()   # give true at place where missing value present
```

```python
df2.isnull().sum()   # show total missing values column-wise
```

```python
df2.isnull().sum().sum()   #total missing values in dataframe
print("Total missing values : ",df2.isnull().sum().sum())
```

```python
df2.shape[0] # shape of column at index 0 give number of rows/values it has
```

```python
# percentage none values in dataframe column-wise
df2.isnull().sum()/df2.shape[0]*100   
```

There are 5 types of merge operations
1. inner join
2. left join | left outer join
3. right join | right outer join
4. Outer join | full outer join
5. cross join

inner join :- join only common data

```python
order = pd.read_excel("Sales Order.xlsx",sheet_name="Orders")
returns = pd.read_excel("Sales Order.xlsx", sheet_name='Returns')
```

```python
print(order.shape,returns.shape)
```

```python
order.head()
```

```python
returns.head()
```

```python
# left dataset, right dataset, left foreign key, right foreign key (left foreign and right foreign key should be present in left and right datasets)

inner_join = pd.merge(left=order,right=returns,left_on="Order ID",right_on="Order ID")  # by default merge do inner join

inner_join.shape
```

```python
# because left_on and right_on dataset are common both side then we can single as 

inner_join = pd.merge(left=order,right=returns, on = "Order ID")
inner_join.shape
```

```python
inner_join.head()
```

If you see in above dataframe , some columns name have x and some have y , here x represent data is from left dataset and y from right

```python
inner_join = pd.merge(left=order,right=returns, on = "Order ID", suffixes=('_orderT',"_returnsT"))
inner_join.head()
```

Left Join :- It means all records from left dataset along with common dataset b/w left and right dataset

```python
returns.rename(columns={"Invoice ID":"Order ID"},inplace=True)
returns.columns
```

```python
left_join = pd.merge(left=order,right=returns, on="Order ID", how='left', indicator=True)
left_join.head()
```

```python
order.shape
```

```python
returns.shape
```

```python
left_join.shape
```

```python
left_join["_merge"].value_counts()
```

Right Join:- It means all records from right dataset along with common dataset b/w left and right dataset

```python
right_join = pd.merge(order,returns,on="Order ID",how='right',indicator=True)
right_join.head()
```

```python
right_join['_merge'].value_counts()
```

```python
returns.shape
```

```python
order.shape
```

```python
right_join.shape
```

Full Join:- In full join, we have all the records of left dataset and right datasets minus common b/w them

```python
full_join = pd.merge(order,returns,on="Order ID",how="outer",indicator=True)
full_join.head()
```

```python
print(order.shape,returns.shape)
```

```python
full_join.shape
```

```python
full_join['_merge'].value_counts()
```

```python
5000+2000
```

```python
7000-89
```

With the help of full join, we can easily find out left join,right join,inner join.

```python
full_join[full_join['_merge']=="both"].head(2)   # inner join
```

```python
full_join[(full_join['_merge']=="both") | (full_join['_merge']=="left_only")].head(2) # left join

# or 

full_join[full_join['_merge'].isin(['both','left_only'])].head(2)
```

```python
full_join[(full_join['_merge']=="both") | (full_join['_merge']=="right_only")].head(2) # right join

# or

full_join[full_join['_merge'].isin(['both','right_only'])].head(2)
```

Cross Join:- It will cross multiply both datasets. That's why there is no need of 'left_on', 'right_on', or 'on' to specify.

```python
cross_join = pd.merge(order[0:5],returns[0:5],how='cross',indicator=True)
cross_join.head(2)
```

```python
cross_join.tail(2)
```

```python
print(order[0:5].shape,returns[0:5].shape)
```

```python
cross_join.shape
```
